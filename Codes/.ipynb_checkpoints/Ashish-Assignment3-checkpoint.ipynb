```python
import numpy as np
import torch
import torch.nn as nn
import sys
import torch.nn.functional as F
import json
from IPython.display import clear_output
from matplotlib import pyplot as plt
import collections
%matplotlib inline
```

```python
def live_plot(data_dict, figsize=(25,10), title=''):
    clear_output(wait=True)
    plt.figure(figsize=figsize)
    for label,data in data_dict.items():
        plt.plot(data, label=label)
    plt.title(title)
    plt.grid(True)
    plt.xlabel('Epoch')
    plt.legend(loc='center left') 
    plt.show();
```

```python
class TEUtils():
      
    def load_glove(glove_path, debug = False):
        f = open(glove_path,'r', encoding='utf8')
        glove_dict = {}
        for line in f:
            token = line.split()
            word = token[0]
            embedding = np.array(list(map(float, token[1:])))
            glove_dict[word] = embedding
            if debug:
                    print('Debug set to True. Set to False when training.')
                    print('Line Read')
                    print('{}'.format(token))
                    print('{}'.format(word))
                    print('{}'.format(embedding))
                    print('{}'.format(glove_dict[word]))

                    break # Prevents accidental printing all the token. Remove when training

        glove_dict['<blank>'] = np.array([0] * 50)
        return glove_dict
    
    def get_embedding_for_word(word, embedding):
        return embedding[word]
    
    def normalize_sentence(sentence, length = 15):
        sentence = sentence.lower().replace('.','').replace('(','').replace(')','').replace("'s'",' is').replace("'re'",' are').replace(',','')
        tokens = sentence.split(' ')
        number_of_words = len(tokens) + 1
        missing_words = length - number_of_words
        for i in range(missing_words + 1):
            sentence = ''.join([sentence, ' <blank>'])
            
        return sentence
            
    
    
    def get_embedding_for_sentence(sentence, embedding, max_length = 15):
        result = []
        for word in TEUtils.normalize_sentence(sentence.lower()).split(' '):
            try:
                embed = TEUtils.get_embedding_for_word(word, embedding)
                result.append(embed)
            except Exception:
                embed = TEUtils.get_embedding_for_word('<blank>', embedding)
                result.append(embed)
            
        return torch.FloatTensor(result).view(15*50)
    
    def get_data(filename, is_test = False):
        val = []
        truth_value = 1
        f = open(filename,'r', encoding='utf8')
        
        for line in f:
            _data = dict()
            json_line = json.loads(line)
            _data['sentence1'] = json_line['sentence1']
            _data['sentence2'] = json_line['sentence2']
            _data['label'] = json_line['annotator_labels']
            neutral = 0
            entailment = 0
            contradiction= 0
            if is_test:
                for label in _data['label']:
                    #label = len(label)
                    if label == 'neutral':
                        neutral += 1
                    if label == 'entailment':
                        entailment += 1
                    if label == 'contradiction':
                        contradiction += 1
                
                if contradiction > neutral:
                    if contradiction > entailment:
                        _data['label'] = [0]
                    else:
                        _data['label'] = [2]
                else:
                    if neutral > entailment:
                        _data['label'] = [1]
                    else:
                        _data['label'] = [2]
                    
            else:
                _data['label'] = _data['label'][0]
                    
            if _data['label'] == 'neutral':
                _data['label'] = [1]
            elif _data['label'] == 'entailment':
                _data['label'] = [2]
            elif _data['label'] == 'contradiction':
                _data['label'] = [0]
                
            _data['label'] = np.array(_data['label'])
            val.append(_data)
            
        
        return val

```

```python
glove_path = r"glove\glove.6B.50d.txt"
```

```python
glove = TEUtils.load_glove(glove_path, debug=False)
blank = glove['<blank>']
glove_dimension = len(blank)
```

```python
class TextEntail(nn.Module):
    def __init__(self, embed_dim, input_size, hidden_size, output_size):
        super(TextEntail, self).__init__()
        #self.input_layer_1 = nn.Linear()
        
        self.dropout = nn.Dropout(p=0.8)
        
        self.hidden_layer_1a = nn.Linear(input_size[0] * embed_dim, hidden_size[0])
        self.output_layer_1a = nn.Linear(hidden_size[0], output_size[0])
        
        self.hidden_layer_1b = nn.Linear(input_size[0] * embed_dim, hidden_size[0])
        self.output_layer_1b = nn.Linear(hidden_size[0], output_size[0])
        
        #self.input_layer_2 = nn.Linear()
        self.hidden_layer_2 = nn.Linear(2 * output_size[0], hidden_size[1])
        self.output_layer_2 = nn.Linear(hidden_size[1], output_size[1])
        
        self.relu = F.relu
        
        
    
    def forward(self, premise, hypothesis):
        # f(n) for premise
        premise_output = self.relu(self.dropout(self.hidden_layer_1a(premise)))
        premise_output = self.relu(self.dropout(self.output_layer_1a(premise_output)))
        
        # f(n) for hypothesis
        hypothesis_output = self.relu(self.dropout(self.hidden_layer_1b(hypothesis)))
        hypothesis_output = self.relu(self.dropout(self.output_layer_1b(hypothesis_output)))
        
        # g(n)
        next_input = torch.cat([premise_output, hypothesis_output])
        output = self.relu(self.dropout(self.hidden_layer_2(next_input)))
        output = self.output_layer_2(output)
        
        return F.softmax(output, dim=-1)
```

```python
model_name = 'model_cross_entropy_100_50.pkt'
use_saved_model = False
model = TextEntail(glove_dimension, [15,100],[500,100],[50,3])
if use_saved_model:
    model.load_state_dict(torch.load(model_name))

print(model)
```

```python
error = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.8)
```

```python
# Training the network
epoch_number = 10000
m = TEUtils.get_data('train.jsonl')

step_print = 500
plot_data = collections.defaultdict(list)
n_data = len(m)
epoch_runs = epoch_number #max(epoch_number, len(m))
sum_loss = 0.0
model.train()
for epoch in range(epoch_runs):
    dataset_point = m[epoch % n_data]
    premise = dataset_point['sentence1']
    hypothesis = dataset_point['sentence2']
    target = torch.LongTensor(dataset_point['label'])
    premise_embedding = TEUtils.get_embedding_for_sentence(premise, glove)
    hypothesis_embedding = TEUtils.get_embedding_for_sentence(hypothesis, glove)
    
    optimizer.zero_grad()
    output = model(premise_embedding, hypothesis_embedding)
    target_class = torch.argmax(target)
    loss = error(output.view(1,3),target)
    sum_loss += loss.item()
    if epoch % step_print == 0:
        plot_data['loss'].append(sum_loss)
        sum_loss = 0.0
        live_plot(plot_data, title='{}/{}'.format(epoch, epoch_runs))
        
    
    loss.backward()
    optimizer.step()
    

    
    
    
```

```python
epoch_number = 200
test_data = TEUtils.get_data('test.jsonl', is_test = True)

step_print = 120
plot_data = collections.defaultdict(list)

pred_correct = 0

epoch_runs = max(epoch_number, len(test_data))
torch.no_grad()
def predict_class(premise, hypothesis):
    premise_embedding = TEUtils.get_embedding_for_sentence(premise, glove)
    hypothesis_embedding = TEUtils.get_embedding_for_sentence(hypothesis, glove)
    
    output = model(premise_embedding, hypothesis_embedding)
    pred_class = torch.argmax(output)
    return output, pred_class

for epoch in range(epoch_runs):
    premise = test_data[epoch]['sentence1']
    hypothesis = test_data[epoch]['sentence2']
    target = torch.LongTensor(test_data[epoch]['label'])
    output, pred_class = predict_class(premise, hypothesis)
    loss = error(output.view(1,3),target)
    target_class = torch.argmax(target)
    print('Premise \t: {}'.format(premise))
    print('Hyporthesis \t: {}'.format(hypothesis))
    if pred_class == target[0]:
        print('RIGHT : Prediction : {} vs Target : {}'.format(pred_class,  target[0]))
        pred_correct += 1
    else:
        print('WRONG : Prediction : {} vs Target : {}'.format(pred_class,  target[0]))
    
    print('----------------------------------------------')
        

    
    
```

```python
print('Accuracy : {}'.format(pred_correct * 100/epoch_runs))
```
